# Visualizing your data

Graphs are a powerful way to communicate the patterns and insights you discover in your data. 

### **Histograms**

   * Show the distribution of numeric data.
   * Bins can be adjusted to change the level of detail.

### **Bar Plots**

   * Compare a categorical variable with numeric values.
   * Often used to display averages or totals by category.

### **Line Plots**

   * Best for showing changes or trends over time.

### **Axis Labels**

   * Can be rotated to improve readability.

### **Scatter Plots**

   * Show relationships between two numeric variables.
   * Useful to spot correlations.

### **Layering Plots**

   * Multiple plots can be displayed together for comparison.

### **Legends & Transparency**

   * Legends identify groups in plots.
   * Transparency (`alpha`) helps overlap visualizations without hiding data.

### **New Dataset (Avocados)**

   * Next examples use avocado sales data, covering categories like size and organic vs. conventional.

## Preparing Data

```python
import pandas as pd
avocados = pd.read_pickle("datasets/avoplotto.pkl")
```

## Exercise: Which avocado size is most popular?

Avocados are widely consumed in the US, and their demand is monitored by the Hass Avocado Board. They categorize sales into three different sizes. In this task, you’ll determine which size has the highest sales using a bar chart.

Bar charts are effective when comparing categories (like avocado size) against numeric values (such as number sold). Before plotting, you’ll need to summarize the data appropriately.

### Instructions

1. Display the first few rows of the `avocados` dataset to review available columns.
2. Group the data by avocado size and calculate the total sales for each size.
3. Create a bar chart to compare the number of avocados sold across sizes.
4. Display the plot.


```python
# Import pyplot for visualization
import matplotlib.pyplot as plt

# Inspect the dataset
print(avocados.head())
```

```python
# Aggregate total sales for each avocado size
size_sales = avocados.groupby("size")["nb_sold"].sum()

# Build the bar chart of sales by size
ax = size_sales.plot.bar(color="skyblue", title="Total Avocado Sales by Size")

# Add axis labels for clarity
ax.set_xlabel("Avocado Size")
ax.set_ylabel("Number Sold")

# Show the plot
plt.show()
```

## Exercise: Changes in sales over time

Line charts are useful for showing how two numeric variables relate when values are connected in sequence. They’re especially handy for seeing how numbers evolve over time, since dates naturally follow one another. In this task, you’ll chart how the number of avocados sold changes across several years.

The dataset contains two records per date—one for organic avocados and one for conventional. You’ll need to combine them before plotting.

### Instructions

1. Calculate the total avocados sold for each date and save it as `nb_sold_by_date`.
2. Build a line chart of these totals over time.
3. Display the chart.

```python
# Sum the number of avocados sold per date
nb_sold_by_date = avocados.groupby("date")["nb_sold"].sum()

# Plot sales trend as a line chart
nb_sold_by_date.plot.line(title="Avocado Sales Over Time", rot=45)

# Add labels for clarity
plt.xlabel("Date")
plt.ylabel("Number of Avocados Sold")

# Display the figure
plt.show()
```

## Exercise: Avocado supply and demand

**Exploring avocado sales and pricing**
Scatter plots are useful when examining how two numerical variables might relate to each other. In this task, you’ll investigate whether the quantity of avocados sold has any connection to their average price. A visible trend could suggest that one variable helps predict the other.

### Instructions

1. Make a scatter plot with the number of avocados sold on the horizontal axis and the average price on the vertical axis.
2. Add a title describing the plot.
3. Display the chart.

```python
# Create a scatter plot: sales vs. price
avocados.plot.scatter(x="nb_sold", y="avg_price")

# Add title and axis labels
plt.title("Avocado Sales vs. Average Price")
plt.xlabel("Number of Avocados Sold")
plt.ylabel("Average Price (USD)")

# Show the scatter plot
plt.show()
```

## Exercise: Price of conventional vs. organic avocados

To better understand price differences, we’ll plot **two histograms** side by side on the same figure. One will show prices of **conventional avocados**, and the other will show **organic avocados**. By using transparency (`alpha`) and increasing the number of bins, we can easily compare how the two distributions overlap and differ.

### Instructions:

1. Subset the dataset for `"conventional"` avocados and plot a histogram of their average prices.
2. Do the same for `"organic"` avocados.
3. Add a **legend** to distinguish between the two groups.
4. Use `alpha=0.5` to make both histograms semi-transparent, so overlaps are visible.
5. Set the number of **bins to 20** for better detail.
6. Add a title and axis labels, then display the plot.

```python
# Histogram of avg_price for conventional and organic avocados
avocados[avocados["type"] == "conventional"]["avg_price"].hist(alpha=0.5, bins=20)
avocados[avocados["type"] == "organic"]["avg_price"].hist(alpha=0.5, bins=20)

# Add legend, labels, and title
plt.legend(["Conventional", "Organic"])
plt.xlabel("Average Price (USD)")
plt.ylabel("Number of Observations")
plt.title("Distribution of Avocado Prices: Conventional vs Organic")

# Display the plot
plt.show()
```
