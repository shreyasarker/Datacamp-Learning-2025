# Annotating time-series data

An effective way to make a visualization more informative is through annotations. These are small pieces of text placed on a chart that highlight specific areas of interest. By pointing out particular details, annotations help direct the viewer’s attention and provide explanations for important features within the data.

### Time-series data

When working with time-series data, the volume of information can sometimes feel overwhelming. Long-term measurements often reveal trends and variations that may span decades. In such cases, annotations can help guide the audience by emphasizing significant points or events within the dataset, making the overall visualization clearer and more meaningful.

### Annotation

Annotations typically combine both text and positioning. A label can be placed at a chosen coordinate on the chart to highlight a specific data value. The text provides context or explanation, while its location ensures it corresponds precisely to the point of interest. Adjusting the placement of the annotation text is often necessary to make it readable and avoid overlapping with other elements of the chart.

### Positioning the text

To improve clarity, the annotation text can be shifted to a more suitable position on the figure. Moving the text away from crowded areas or axis labels ensures that the message is visible and easy to interpret. However, repositioning the text introduces the need to visually connect it with the data point it refers to.

### Adding arrows to annotation

This connection is usually established by adding an arrow that links the annotation text to the data point. The arrow makes it immediately clear which value is being described, reducing any ambiguity.

### Customizing arrow properties

The appearance of the arrow can also be adjusted to suit the overall style of the visualization. Properties such as line type, head shape, and color can be customized, helping to make the annotation stand out while keeping the chart visually balanced.

### Customizing annotations

Beyond arrow styles, many other aspects of annotations can be customized. These include the placement, font, and alignment of the text, as well as additional stylistic details. Matplotlib provides a wide range of options for tailoring annotations to create clear, professional, and effective visualizations.


## Prepare Data

```python
# Import pandas library
import pandas as pd

# Read the CSV file and set up a DateTime index
climate_change = pd.read_csv(
    "datasets/climate_change.csv",
    parse_dates=["date"],
    index_col="date"
)
```

## Exercise: Annotating a plot of time-series data

Annotations are helpful for calling out key insights directly on a plot. They allow us to point to specific data points and provide explanatory text. For instance, when analyzing the climate change dataset, we might want to emphasize the moment when the relative temperature first crossed 1 degree Celsius.

In this task, you will use the `annotate` method from Matplotlib’s Axes object. The dataset `climate_change` is already available as a DataFrame with dates as its index. You’ll plot the relative temperature over time and add an annotation to highlight the important point.

### Instructions

* Use `ax.plot` to display the `relative_temp` column against the DataFrame index.
* Add an annotation with the label `'>1 degree'` at the coordinates `(pd.Timestamp('2015-10-06'), 1)`.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Plot relative temperature over time
ax.plot(climate_change.index, climate_change["relative_temp"], color="red")

# Annotate the first date when relative temperature exceeded 1 degree Celsius
ax.annotate(
    text=">1 degree",
    xy=(pd.Timestamp("2015-10-06"), 1)
)

plt.xlabel("Time (years)")
plt.ylabel("Relative temperature (Celsius)")
plt.title("Climate Change: Relative Temperature")
plt.show()
```
