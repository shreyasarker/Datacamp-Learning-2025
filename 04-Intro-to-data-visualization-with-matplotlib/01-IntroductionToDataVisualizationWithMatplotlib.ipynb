# Introduction to Data Visualization with Matplotlib
Data visualizations help uncover insights from data and provide a way to communicate those insights effectively. Visual representations often convey information more clearly than raw numbers alone.

### Data Visualization
Matplotlib is a widely-used Python library for creating data visualizations. Its strength lies in giving you precise control over the appearance and properties of your plots. By the end of the course, you will be able to not only create and customize visualizations but also automate their creation based on your data.

### Introducing the Pyplot Interface
Matplotlib can be used in several ways, and this course focuses on its object-oriented interface via the pyplot submodule. Using pyplot, you can create a Figure object (the container for the visualization) and an Axes object (the area where data is drawn). This setup provides a flexible structure for building and customizing visualizations.

### Adding Data to Axes
Data is added to the Axes object using plotting commands. These commands take the data and map it onto the visual space of the Axes, allowing you to transform raw numbers into visual patterns that are easier to interpret.

### Plotting Data
By calling plotting methods on the Axes object and displaying the result, you can create visual representations of your data. Trends and patterns become easier to recognize compared to analyzing raw numbers in a table.

### Adding More Data
Multiple datasets can be added to the same Axes object, enabling comparison and more comprehensive visual analysis. Each dataset is added through additional plotting commands.

### Putting It All Together
The full workflow involves creating the Figure and Axes objects, plotting one or more datasets on the Axes, and finally rendering the figure. This structured approach allows for clear, flexible, and customizable visualizations.

## Prepare Data

Matplotlib can be used in multiple ways, but this exercise focuses on the `pyplot` interface, which is highly flexible for creating and customizing visualizations. Using `pyplot`, you can generate two primary objects: a Figure (the overall container) and an Axes (the area where data is plotted).

## Exercise: Adding data to an Axes object

To visualize data in Matplotlib, you add it to an Axes object using its methods. In this exercise, you'll compare monthly rainfall in two U.S. cities: Seattle and Austin.

The datasets `seattle_weather` and `austin_weather` are already loaded as pandas DataFrames. Both contain a `"MONTH"` column for the month names and a `"MLY-PRCP-NORMAL"` column for average monthly rainfall over a ten-year period. Also, matplotlib.pyplot as plt is also imported.

### Instructions

1. Create a Figure and Axes object using `plt.subplots()`.
2. Plot Seattle's monthly rainfall on the Axes using the `plot` method.
3. Plot Austin's monthly rainfall similarly.
4. Display the figure using `plt.show()`.

```python
# Generate a Figure and Axes object
fig, ax = plt.subplots()

# Add Seattle rainfall data to the Axes
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], label="Seattle")

# Add Austin rainfall data to the Axes
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], label="Austin")

# Display the figure
plt.show()
```
