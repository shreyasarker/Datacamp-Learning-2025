# Quantitative comparisons: bar-charts

In earlier discussions, we focused on turning data into visual forms. Now, the emphasis shifts toward making quantitative comparisons across different parts of a dataset.

When dealing with numerical values across categories, raw tables often make it hard to compare values quickly. A visual approach, such as a bar chart, can make these comparisons more intuitive by showing differences in magnitude directly.

To begin, data can be loaded into a DataFrame, using meaningful identifiers like category names as the index. A simple bar chart can then be created, where the bar height corresponds to the value of interest, and the labels along the x-axis correspond to the categories. However, long labels can create clutter, which can be improved by adjusting their orientation or rotation. Adding axis labels further clarifies the meaning of the visualization.

Beyond showing a single variable, multiple related variables can also be represented. A stacked bar chart allows multiple quantities to be displayed in one combined visualization. Each new variable is stacked on top of the previous, so the total bar height represents a combined value while still showing the contribution of each individual component.

To ensure clarity, legends are used to identify which color or section corresponds to which variable. This involves assigning labels during plotting and then displaying them with a legend command. With these additions, the final stacked bar chart becomes both comprehensive and easy to interpret, allowing quick comparisons across categories and variables.

## Prepare Data

```python
# Import pandas library
import pandas as pd

# Read the CSV file and set up a DateTime index
medals = pd.read_csv(
    "datasets/medals_by_country_2016.csv",
    index_col=0
)
```

## Exercise: Bar chart

Bar charts are useful for showing how values differ across categories. Each bar’s height represents the value for a specific category, making it easier to compare groups visually.

In this exercise, you’ll work with a DataFrame named `medals`. The index represents different countries, while the `"Gold"` column stores the number of gold medals earned by each. You’ll display these values in a bar chart.

### Instructions

* Use the `ax.bar` method to create a bar chart of the `"Gold"` column against the countries.
* Adjust the x-axis tick labels so they display the country names.
* Rotate these tick labels by 90 degrees for readability.
* Label the y-axis as **"Number of medals"**.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Create a bar chart for gold medals per country
ax.bar(medals.index, medals["Gold"])

# Rotate and set country names as tick labels
ax.set_xticks(range(len(medals.index)))
ax.set_xticklabels(medals.index, rotation=90)

# Label the y-axis
ax.set_ylabel("Number of medals")

plt.show()
```

## Exercise: Stacked bar chart

Stacked bar charts allow us to compare totals while also breaking them down into categories. Each bar shows the combined height of multiple variables, with each variable stacked on top of the previous one. This makes it possible to see both the overall total and the contribution of each category.

In this exercise, you’ll use a DataFrame called `medals`. The index represents countries, and the columns `"Gold"`, `"Silver"`, and `"Bronze"` represent the number of medals of each type. You’ll create a stacked bar chart showing how many of each medal type each country has won.

### Instructions

* Add a bar chart for the `"Gold"` medals, setting the label to **"Gold"**.
* Stack the `"Silver"` bars on top of the gold bars using the `bottom` argument, and label them **"Silver"**.
* Add the `"Bronze"` bars on top of both, again using the `bottom` argument, and label them **"Bronze"**.
* Include a legend so the chart clearly indicates which color corresponds to which medal type.

```python
fig, ax = plt.subplots()

# Plot the gold medal bars
ax.bar(medals.index, medals["Gold"], label="Gold")

# Stack silver medals on top of gold
ax.bar(medals.index, medals["Silver"],
       bottom=medals["Gold"], label="Silver")

# Stack bronze medals on top of gold + silver
ax.bar(medals.index, medals["Bronze"],
       bottom=medals["Gold"] + medals["Silver"], label="Bronze")

# Rotate and set country names as tick labels
ax.set_xticks(range(len(medals.index)))
ax.set_xticklabels(medals.index, rotation=90)

# Label the y-axis
ax.set_ylabel("Number of medals")

# Add a legend for clarity
ax.legend()

plt.show()
```
