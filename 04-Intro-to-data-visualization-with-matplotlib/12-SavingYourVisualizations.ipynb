# Saving your visualizations

Once a visualization has been created, the next step is preparing it to be shared. This involves final adjustments to the figure and saving it in a format that is suitable for collaboration or distribution.

When generating a plot, it typically appears on the screen during development. However, instead of displaying it, you can save the figure as a file directly. This allows you to store it in your working directory and then share it with others.

The format in which a figure is saved depends on its intended use. Some formats preserve image quality at the cost of larger file sizes, while others reduce file size by compressing the image. Certain formats also support further editing of individual elements in external graphics tools. Choosing the right format is part of preparing the figure for its final purpose.

Another factor to consider is resolution. The clarity of the image is controlled by the number of dots rendered per inch. A higher resolution produces sharper images but also increases file size. The balance between quality and storage efficiency depends on the context in which the visualization will be used.

In addition to resolution, the physical size of the figure can also be customized. By adjusting its width and height, you control not only how large the figure appears but also its aspect ratio. This allows flexibility in presenting the figure in different layouts, whether for digital use or printed materials.


## Prepare Data

```python
# Import pandas library
import pandas as pd

# Read the CSV file
medals = pd.read_csv(
    "datasets/medals_by_country_2016.csv",
    index_col=0
)
```

## Exercise: Saving a file several times

When you want to share charts or graphs with others, saving them as image files is essential. Matplotlib makes this possible with the `savefig()` method. In this task, you will practice exporting the same figure more than once, while changing some of the saving options each time. A figure has already been created for you.

### Instructions

1. First, display the figure on the screen using `plt.show()`.
2. Save the figure to a file called **chart.png** with the standard resolution.
3. Save the same figure again, this time calling the file **chart\_highres.png**, and set the resolution to **300 dpi**.

```python
import matplotlib.pyplot as plt

# Create the figure and axes
fig, ax = plt.subplots()

# Draw a bar plot showing gold medals by country
ax.bar(medals.index, medals["Gold"])

# Adjust the tick labels for readability
ax.set_xticks(range(len(medals.index)))
ax.set_xticklabels(medals.index, rotation=90)

# Add a label to the y-axis
ax.set_ylabel("Gold medals won")

# Step 1: Show the figure
plt.show()
```

```python
# Step 2: Save figure with default settings
fig.savefig("datasets/chart.png")
```

```python
# Step 3: Save figure again with higher resolution
fig.savefig("datasets/chart_highres.png", dpi=300)
```

## Exercise: Save a figure with different sizes

Sometimes, before saving your visualization, you may want to control its physical size on the page or screen. Matplotlib allows this through the `set_size_inches()` method of a `Figure` object. This method requires two numbers: the first defines the width, and the second defines the height.

In this activity, you will use the provided figure (`fig`) and save it in two versions, each with a different size.

### Instructions

1. Adjust the figure size to **3 inches wide** and **5 inches tall**, then save it as **figure\_3\_5.png** using default settings.
2. Change the figure size to **5 inches wide** and **3 inches tall**, and save it again as **figure\_5\_3.png** with default resolution.

```python
# Create the figure and axes
fig, ax = plt.subplots()

# Draw a bar plot showing gold medals by country
ax.bar(medals.index, medals["Gold"])

# Adjust the tick labels for readability
ax.set_xticks(range(len(medals.index)))
ax.set_xticklabels(medals.index, rotation=90)

# Add a label to the y-axis
ax.set_ylabel("Gold medals won")

# Step 1: Set size to 3x5 inches and save
fig.set_size_inches(3, 5)
fig.savefig("datasets/figure_3_5.png")
```

```python
# Step 2: Set size to 5x3 inches and save
fig.set_size_inches(5, 3)
fig.savefig("datasets/figure_5_3.png")
```
