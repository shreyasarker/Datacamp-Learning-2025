# Preparing your figures to share with others

This chapter emphasizes how to make visualizations ready for sharing and integration into automated analysis workflows. The focus begins with figure style customization. Previously, adjustments were made to individual elements, such as changing line colors or marker shapes. Now, the scope expands to altering the overall appearance of a figure.

### Changing plot style
Instead of modifying each element separately, you can adjust the entire style of a figure at once. This changes the overall look and feel of the visualization, affecting multiple components simultaneously.

### Choosing a style
When a new style is applied before plotting, the figure undergoes a complete transformation. The changes go beyond one visual element, influencing colors, fonts, and background settings. Gridlines and other features may also appear, giving the visualization a different overall aesthetic. Importantly, once a style is selected, it remains active for all subsequent plots within the same session until replaced with another style.

### Back to the default
If needed, the style can be reset to the default option with a simple command. This restores the original appearance used by Matplotlib when no custom style is specified.

### The available styles
Matplotlib includes several built-in styles. A complete list can be found in its documentation, where each style is demonstrated with example visualizations.

### The "bmh" style
Among the options, one available choice is the style labeled “bmh,” which changes the figure’s appearance according to its predefined settings.

### Seaborn styles
Several styles originate from the Seaborn library, which was developed for statistical visualization. Matplotlib has incorporated some of these styles, and they are accessible by name. These provide alternative aesthetics designed with clear communication and statistical graphics in mind.

### Guidelines for choosing plotting style
When deciding which style to use, consider the audience and the context. Dark backgrounds are often avoided because they reduce readability. Colorblind-friendly options are highly recommended, since a significant portion of people cannot distinguish certain colors. The medium of presentation also matters: figures intended for websites can differ from those prepared for print. If the figures are likely to be printed, using styles that minimize ink usage is preferable. In cases where black-and-white printing is expected, grayscale styles ensure that distinctions between elements remain visible.

## Prepare Data

```python
# Import pandas library
import pandas as pd

# Read the CSV file
seattle_weather = pd.read_csv("datasets/seattle_weather_simple.csv")
austin_weather = pd.read_csv("datasets/austin_weather_simple.csv")
```

## Excercise: Switching between styles

Switching the style of plots can change how all subsequent visualizations appear.
In this activity, you’ll experiment with applying two different styles while plotting average monthly temperatures. You’ll be working with two datasets: **seattle\_weather** and **austin\_weather**, both containing monthly records of average temperatures.

### Instructions

1. Apply the **"ggplot"** theme, then create a new figure and axis, and display the Seattle monthly average temperatures.
2. Change the style to **"seaborn-colorblind"**, then create a new figure and axis, and plot Austin monthly average temperatures.

```python
import matplotlib.pyplot as plt

# Apply the ggplot theme and make a fresh figure/axis
plt.style.use("ggplot")
fig, ax1 = plt.subplots()
ax1.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], marker="o")
ax1.set_title("Seattle Monthly Avg Temp - ggplot Style")
plt.show()
```

```python
# Switch to seaborn-colorblind style and plot Austin data
plt.style.use("Solarize_Light2")
fig, ax2 = plt.subplots()
ax2.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"], marker="s")
ax2.set_title("Austin Monthly Avg Temp - Solarize_Light2 Style")
plt.show()
```
