# Quantitative comparisons: histograms

Bar charts allow us to compare the values of a variable across different conditions. However, histograms provide a deeper insight because they display the full distribution of values within a variable rather than just a summary measure.

When working with data, we may want to examine attributes such as height or weight across groups. While a bar chart can give us the average values for comparison, it does not reveal the spread or variation. In contrast, histograms present the complete distribution, showing how values are spread across intervals.

A histogram uses the x-axis to represent the range of values within the variable, while the y-axis shows the frequency of observations within specific intervals. This enables us to see not only where values are concentrated but also how they differ across groups.

Since colors alone do not clarify which dataset corresponds to which histogram, labels and legends are essential. Adding them ensures that viewers can easily distinguish between the groups being compared.

By default, a histogram divides the data into ten intervals or bins. This can be customized by specifying a different number of bins, allowing for finer or broader divisions of the data. Alternatively, exact bin boundaries can be defined, giving control over where the splits occur.

In some cases, overlapping histograms may obscure important details. This issue can be resolved by adjusting the visualization style. For instance, instead of solid bars, using line-based representations improves clarity by removing the occlusion effect.

With these customizations, histograms not only provide a clear view of distributions but also allow more flexibility in how the data is displayed and interpreted.

## Prepare Data

```python
# Import pandas library
import pandas as pd

# Read the CSV file and set up a DateTime index
mens_rowing = pd.read_csv("datasets/mens_rowing.csv")
mens_gymnastics = pd.read_csv("datasets/mens_gymnastics.csv")
```

## Exercise: Creating histograms

Histograms let us see the entire distribution of a variable rather than just an average. In this task, you’ll compare the distribution of weights of medalists from the **2016 Olympic Games** in two sports: men’s rowing and men’s gymnastics.

You are given two DataFrames:

* **`mens_rowing`** → medalists in men’s rowing events
* **`mens_gymnastics`** → medalists in men’s gymnastics events

Your goal is to create histograms that show the distribution of weights for both groups.

### Instructions

1. Use `ax.hist()` to plot the **"Weight"** column from the `mens_rowing` DataFrame.
2. Add another histogram for the **"Weight"** column from the `mens_gymnastics` DataFrame.
3. Label the x-axis as **"Weight (kg)"**.
4. Label the y-axis as **"# of observations"**.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Histogram of weights for men's rowing
ax.hist(mens_rowing["Weight"])

# Histogram of weights for men's gymnastics
ax.hist(mens_gymnastics["Weight"])

# Set axis labels
ax.set_xlabel("Weight (kg)")
ax.set_ylabel("# of observations")

# Show the figure
plt.show()
```

## Exercise: "Step" histogram

Histograms let us compare the distributions of data between different groups. In this exercise, you will look at the **weights of medalist athletes** in two sports from the **2016 Summer Olympics**: **Rowing** and **Gymnastics**.

The data is provided in two DataFrames:

* `mens_rowing` → medalists in men’s rowing events
* `mens_gymnastics` → medalists in men’s gymnastics events

You will create **step-style histograms** to visualize and compare their weight distributions.

### Instructions

1. Plot a histogram of the `"Weight"` column from `mens_rowing`.
   * Use the label **"Rowing"**.
   * Set the histogram type (`histtype`) to **"step"**.
   * Use **5 bins**.

2. Plot another histogram of the `"Weight"` column from `mens_gymnastics`.
   * Use the label **"Gymnastics"**.
   * Also set `histtype="step"` and `bins=5`.

3. Add **x-axis** label → `"Weight (kg)"`.
4. Add **y-axis** label → `"# of observations"`.
5. Add a **legend** before displaying the figure.

```python
fig, ax = plt.subplots()

# Histogram for rowing medalists
ax.hist(mens_rowing["Weight"], 
        label="Rowing", 
        histtype="step", 
        bins=5)

# Histogram for gymnastics medalists
ax.hist(mens_gymnastics["Weight"], 
        label="Gymnastics", 
        histtype="step", 
        bins=5)

# Axis labels
ax.set_xlabel("Weight (kg)")
ax.set_ylabel("# of observations")

# Add legend
ax.legend()

# Show figure
plt.show()
```
