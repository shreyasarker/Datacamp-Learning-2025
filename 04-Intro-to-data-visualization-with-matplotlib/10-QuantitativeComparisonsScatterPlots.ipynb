# Quantitative comparisons: scatter plots

Bar charts display the values of one variable across different categories, such as countries. However, when the goal is to compare two variables across observations, a different type of visualization is needed. This is referred to as a bi-variate comparison, since it examines the relationship between two variables simultaneously.

### Introducing scatter plots
A common visualization for bi-variate comparisons is the scatter plot. In this type of plot, each observation is represented as a point. The position of the point along the x-axis corresponds to the value of one variable, while the position along the y-axis represents the value of the second variable. To create this visualization, a Figure and Axes object are initialized, and the `scatter` method of the Axes is used. The method takes the variable for the x-axis as the first argument and the variable for the y-axis as the second argument. Axis labels are then added to clarify interpretation, and the figure is displayed.

### Customizing scatter plots
Scatter plots can be customized in the same way as other visualizations. When displaying more than one set of bi-variate comparisons on the same axes, customization ensures the plots remain distinct. Different data selections can be added to the same Axes object, and visual adjustments such as point color and labeling help differentiate them. Finally, a legend is included to make the distinction between datasets clear, alongside axis labels and display of the figure.

### Encoding a comparison by color
Color can be used as a tool for comparison within scatter plots. It allows one to distinguish between different groups or conditions directly on the same figure, making the relationship between variables easier to interpret.

### Encoding a third variable by color
Beyond simple group comparisons, color can also encode an additional variable. Instead of assigning a fixed color, the `c` keyword argument enables the values of another variable to determine the coloring of points. This approach integrates more information into a single visualization while still focusing on the relationship between the two primary variables.

### Encoding time in color
When time is encoded as color, the shade or brightness of the points reflects when the measurement was taken. Earlier observations may appear in darker tones, while later observations are shown in lighter or brighter tones, providing a temporal perspective within the scatter plot.

## Prepare Data

```python
# Import pandas library
import pandas as pd

# Read the CSV file and set up a DateTime index
climate_change = pd.read_csv(
    "datasets/climate_change.csv",
    parse_dates=["date"],
    index_col="date"
)
```

## Exercise: Simple scatter plot

Scatter plots are a way to visualize the relationship between two variables. Each row of data is represented as a point on the graph. The value of the first variable determines its position along the horizontal axis, while the value of the second variable determines its position along the vertical axis.

In this task, you’ll draw a scatter plot using the `climate_change` DataFrame. This dataset already contains a column `"co2"`, which records monthly carbon dioxide levels, and another column `"relative_temp"`, which shows the temperature at the same time.

### Instructions

* Use the `scatter()` function from the Axes object to plot `"co2"` on the x-axis and `"relative_temp"` on the y-axis.
* Label the x-axis as **"CO2 (ppm)"**.
* Label the y-axis as **"Relative Temperature (°C)"**.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Create scatter plot with CO2 on the x-axis and relative temperature on the y-axis
ax.scatter(x=climate_change["co2"], y=climate_change["relative_temp"])

# Label the axes
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative Temperature (°C)")

# Render the plot
plt.show()
```

## Exercise: Encoding time by color

Although a scatter plot is drawn in two dimensions, it is possible to represent an extra piece of information by using color. In this activity, you will create a scatter plot from the `climate_change` dataset. The `"co2"` values will be plotted along the horizontal axis, and the `"relative_temp"` values will be plotted along the vertical axis. The additional dimension of time will be represented through color, where earlier observations will appear in darker blue shades and more recent ones will appear in brighter yellow shades.

### Instructions

* Plot `"co2"` against `"relative_temp"` using the `scatter()` method of the Axes object.
* Use the `c` parameter to map the DataFrame’s index to color, so that the timeline is visually encoded.
* Label the x-axis as **"CO2 (ppm)"** and the y-axis as **"Relative Temperature (°C)"**.

```python
fig, ax = plt.subplots()

# Create scatter plot: CO2 vs relative temperature, color mapped to time index
ax.scatter(
    x=climate_change["co2"],
    y=climate_change["relative_temp"],
    c=climate_change.index
)

# Label axes
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative Temperature (°C)")

# Display the plot
plt.show()
```
