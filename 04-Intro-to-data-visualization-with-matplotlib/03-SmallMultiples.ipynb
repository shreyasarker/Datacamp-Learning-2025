# Small multiples

Sometimes, adding more information to a single plot makes it cluttered and harder to interpret. For instance, we may begin with one dataset, like average precipitation, and then add details such as percentiles or even data from another location. However, combining everything into one plot can quickly become overwhelming. A better approach in such cases is to use *small multiples*—separate but related plots that allow for easier comparisons across datasets.

In Matplotlib, these are created using **subplots**, which organize multiple plots into a grid of rows and columns. When multiple subplots are created, they form an array of Axes objects, and we can add data to each subplot by indexing into this array. If only one row or column of plots is generated, the array is one-dimensional, requiring only a single index to access each subplot. This structure makes it simple to assign different datasets to different plots, while customizing labels for clarity.

By arranging data across multiple subplots, comparisons become much clearer, although one issue that may arise is inconsistent axis ranges between plots. To resolve this, we can use the **sharey=True** option when creating subplots, ensuring that all plots share the same y-axis range. This provides a more consistent basis for comparison and makes patterns across datasets easier to see.

## Prepare Data

```python
import pandas as pd

seattle_weather = pd.read_csv("datasets/seattle_weather_simple.csv")
austin_weather = pd.read_csv("datasets/austin_weather_simple.csv")
```

## Exercise: Creating small multiples with plt.subplots

Small multiples allow us to compare several datasets in a clear and organized way by displaying them side-by-side. In Matplotlib, these can be created with the `plt.subplots()` function, where the first argument specifies the number of rows and the second defines the number of columns.

In this task, you’ll use the weather data from Seattle and Austin. Each DataFrame (`seattle_weather` and `austin_weather`) includes the columns `"MONTH"`, `"MLY-PRCP-NORMAL"` (monthly average rainfall), and `"MLY-TAVG-NORMAL"` (monthly average temperature). You will display precipitation and temperature for both cities, arranging them in a 2-by-2 grid of plots.

### Instructions

* Generate a figure with a 2×2 grid of subplots.
* Plot Seattle’s precipitation in the top-left (row 0, column 0).
* Plot Seattle’s average temperature in the top-right (row 0, column 1).
* Plot Austin’s precipitation in the bottom-left (row 1, column 0).
* Plot Austin’s average temperature in the bottom-right (row 1, column 1).

```python
# Import pyplot from matplotlib
import matplotlib.pyplot as plt

# Create a figure with 2 rows and 2 columns of subplots
fig, axes = plt.subplots(2, 2)

# Top-left: Seattle precipitation
axes[0, 0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])

# Top-right: Seattle temperature
axes[0, 1].plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"])

# Bottom-left: Austin precipitation
axes[1, 0].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"])

# Bottom-right: Austin temperature
axes[1, 1].plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])

# Display the entire figure
plt.show()
```

## Exercise: Small multiples with shared y axis

When working with small multiples, it’s often helpful to keep the y-axis scale consistent across plots. This makes it easier to compare datasets visually. You can do this in Matplotlib by setting the `sharey` argument to `True` when creating subplots.

In this task, you’ll build a figure with two plots stacked vertically, both sharing the same y-axis. You will plot Seattle’s precipitation data in the top panel and Austin’s precipitation data in the bottom panel. Each will include average values and the 25th and 75th percentiles for comparison.

### Instructions

* Create a figure with two subplots arranged in one column, sharing the y-axis.
* On the top plot, draw Seattle’s average precipitation in a solid blue line, and its 25th and 75th percentiles in dashed blue lines.
* On the bottom plot, show Austin’s average precipitation in a solid red line, with the 25th and 75th percentiles in dashed red lines.

```python
# Build a figure with 2 rows and 1 column, sharing the y-axis
fig, axes = plt.subplots(nrows=2, ncols=1, sharey=True)

# Top plot: Seattle precipitation
axes[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"], color="blue")
axes[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-25PCTL"], color="blue", linestyle="--")
axes[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-75PCTL"], color="blue", linestyle="--")

# Bottom plot: Austin precipitation
axes[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], color="red")
axes[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-25PCTL"], color="red", linestyle="--")
axes[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-75PCTL"], color="red", linestyle="--")

# Display the figure
plt.show()
```
