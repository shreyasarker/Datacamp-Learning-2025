# Customizing your plots
Now that you’ve learned how to plot data, let’s explore how to make your visualizations more customized.

### Customizing data appearance
One way to improve your plots is by adjusting how the data looks. For example, although data may be displayed as continuous, sometimes it represents measurements taken at intervals. You can make this clearer by customizing its appearance.

### Adding markers
Markers can be added to highlight individual data points. This makes it clear where actual data exists and which parts are simply connecting lines.

### Choosing markers
You have a variety of marker styles to choose from, each representing data points in a different way. Selecting a marker type makes the plot more readable and emphasizes measured values.

### Setting the linestyle
You can also change how the connecting lines look by adjusting the line style. For example, they can appear as dashed, dotted, or in other formats, depending on what you want to emphasize.

### Eliminating lines with linestyle
It’s even possible to remove the connecting lines entirely, leaving only the data points visible.

### Choosing color
You can modify the color of the data in the plot to make it more distinct or visually appealing.

### Customizing the axes labels
Labeling your axes is crucial for clear communication. The Axes object provides methods to set axis labels, ensuring viewers understand what each dimension of the plot represents. Proper capitalization should also be used for readability.

### Setting the y-axis label
Similar methods exist for labeling the y-axis, allowing you to clarify what the vertical values represent.

### Adding a title
You can also add a title to the plot, providing context and making the purpose of the visualization clearer.


## Prepare Data

```python
import pandas as pd

seattle_weather = pd.read_csv("datasets/seattle_weather_simple.csv")
austin_weather = pd.read_csv("datasets/austin_weather_simple.csv")
```

## Exercise: Customizing data appearance

When adding data to plots, you can modify how it appears by using keyword arguments within the plotting command.

In this task, you will adjust the marker style, line style, and colors for data from two cities.

The data is stored in pandas DataFrames already available: `seattle_weather` and `austin_weather`. Each contains a `"MONTH"` column and a `"MLY-PRCP-NORMAL"` column.

### Instructions

* Use `ax.plot` to plot `"MLY-PRCP-NORMAL"` against `"MONTH"` for both DataFrames.
* Customize the Seattle data with a blue line, circle markers, and dashed style.
* Customize the Austin data with a red line, downward triangle markers, and dashed style.

```python
# Import pyplot from matplotlib
import matplotlib.pyplot as plt

# Generate a Figure and Axes object
fig, ax = plt.subplots()

# Plot data for Seattle with custom style
ax.plot(
    seattle_weather["MONTH"], 
    seattle_weather["MLY-PRCP-NORMAL"], 
    color="blue", marker="o", linestyle="--"
)

# Plot data for Austin with custom style
ax.plot(
    austin_weather["MONTH"], 
    austin_weather["MLY-PRCP-NORMAL"], 
    color="red", marker="v", linestyle="--"
)

# Display the plot
plt.show()
```

## Exercise: Customizing axis labels and adding titles

To customize axis labels in Matplotlib, you can use the `set_xlabel()` and `set_ylabel()` methods of an Axes object. Similarly, you can add a plot title using `set_title()`.

In this task, you’ll label both axes and give the plot a descriptive title.

The data is already prepared in pandas DataFrames: `seattle_weather` and `austin_weather`. Each contains `"MONTH"` and `"MLY-PRCP-NORMAL"` columns.
A `Figure` object (`fig`) and an `Axes` object (`ax`) have also been created.

### Instructions

* Add `"Time (months)"` as the x-axis label.
* Add `"Precipitation (inches)"` as the y-axis label.
* Add `"Weather trends in Austin and Seattle"` as the title.

```python
# Generate a Figure and Axes object
fig, ax = plt.subplots()

# Plot precipitation data for both cities
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"])

# Set x-axis label
ax.set_xlabel("Time (months)")

# Set y-axis label
ax.set_ylabel("Precipitation (inches)")

# Set the plot title
ax.set_title("Weather trends in Austin and Seattle")

# Show the final plot
plt.show()
```
