# Plotting time-series data

Time-series data is a common format where information is recorded over time, making it particularly useful for identifying trends and patterns. Variables such as temperature or rainfall are often organized according to time intervals, like days, months, or years. When working with time-series data in pandas, it’s important to ensure that date columns are properly parsed and, ideally, set as the DataFrame index. This creates a DateTimeIndex, which Matplotlib can recognize and handle effectively for plotting. Alongside the date index, the dataset may include other continuous variables, often stored as floating-point numbers, which can be used for analysis. Occasionally, gaps in data may occur, recorded as missing values (NaNs), which appear as breaks in visualizations. Plotting such data involves using the index for the x-axis and the variable of interest for the y-axis, with Matplotlib automatically formatting time labels on the axis, such as years or months depending on the scale. By slicing subsets of the data, it is possible to zoom in on specific periods, like a decade or a single year, with the axis adjusting automatically to reflect appropriate intervals, whether by years or months. This flexibility makes time-series visualizations powerful for both broad overviews and detailed inspections.


## Exercise: Read data with a time index

Pandas allows DataFrames to be indexed by dates, which Matplotlib automatically recognizes for time-series plots.

In this exercise, we’ll load the file **climate\_change.csv**, which includes monthly records of CO2 emissions and temperature variations between 1958 and 2016. By parsing the `"date"` column as a datetime object and assigning it as the index, we can work seamlessly with time-based data.

Keep the **Matplotlib Cheat Sheet** nearby in case you need a quick review!

### Instructions

1. Import pandas and give it the alias `pd`.
2. Use `pd.read_csv` to read the CSV file.
3. Parse the `"date"` column as datetime values.
4. Make the `"date"` column the index of the DataFrame.

```python
# Import pandas library
import pandas as pd

# Read the CSV file and set up a DateTime index
climate_df = pd.read_csv(
    "datasets/climate_change.csv",
    parse_dates=["date"],
    index_col="date"
)
```

## Exercise: Plot time-series data

When working with time-based datasets, plotting is usually done by taking the index (dates) for the x-axis and a variable of interest for the y-axis. The `plot()` method from the Axes object allows us to map these values directly.

Here, you’ll use the `climate_change` DataFrame, which is already indexed by date. It contains two key columns:

* `"co2"` (carbon dioxide levels)
* `"relative_temp"` (temperature anomaly values).

Your task is to visualize how `"relative_temp"` changes over time and add meaningful axis labels.

### Instructions

1. Create a figure and an Axes using `plt.subplots()`.
2. Plot the `"relative_temp"` column against the time index.
3. Label the x-axis as **Time**.
4. Label the y-axis as **Relative temperature (Celsius)**.
5. Display the plot.

```python
import matplotlib.pyplot as plt

# Set up the figure and axes
figure, axis = plt.subplots()

# Plot relative temperature values over time
axis.plot(climate_df.index, climate_df["relative_temp"], color="purple")

# Add descriptive labels
axis.set_xlabel("Time")
axis.set_ylabel("Relative temperature (Celsius)")

# Render the plot
plt.show()
```

## Exercise: Using a time index to zoom in

When working with time-series data, we can take advantage of the time index to look at specific time ranges. By selecting only the rows that fall within a given period, we can zoom in and analyze trends more clearly. In this exercise, we’ll focus on CO₂ levels during the 1970s and visualize them.

### Instructions

1. Import `matplotlib.pyplot` as `plt`.
2. Create a figure and axes using `plt.subplots()`.
3. Select the data between **January 1, 1970** and **December 31, 1979** from the `climate_change` DataFrame.
4. Plot the `"co2"` values against the time index for this period.
5. Display the plot with `plt.show()`.

```python
# Create a figure and axis
fig, ax = plt.subplots()

# Extract data for the 1970s
seventies = climate_df.loc["1970-01-01":"1979-12-31"]

# Plot CO2 levels during the 1970s
ax.plot(seventies.index, seventies["co2"], color="blue", linestyle="--")

# Show the plot
plt.show()
```
