## Statistical plotting 

In the earlier lesson, histograms were introduced as a way to visualize and compare data distributions. However, comparisons can also be made in a more structured manner. Statistical plotting refers to a collection of visualization techniques designed for making such comparisons clearer and more formal. Two important methods in this category will be discussed here.

One technique involves adding error bars to charts. Error bars act as additional markers that provide insight into the variability or spread of the data. While histograms reveal the entire distribution, error bars condense this into a summary, often using a measure like the standard deviation. When applied to bar charts, the error bar is included as an argument that supplements the mean value with an indication of the variability in the data.

Error bars are not limited to bar charts; they can also be incorporated into line plots. By using them in this context, both the central tendency and the variation around it are communicated. The vertical markers serve as a quick reference to understand how consistent or spread out the values are across the data points.

Another technique for statistical plotting is the boxplot. This visualization provides multiple layers of information within a single compact figure. It highlights central values, the spread of the middle range of the data, and the overall extent of the distribution.

A boxplot is particularly powerful because it not only summarizes the data through medians and quartiles but also draws attention to variability and unusual values. It makes it possible to quickly identify the range in which most data falls, while also detecting values that fall far outside the expected pattern, often considered outliers.

## Prepare Data

```python
# Import pandas library
import pandas as pd

# Read the CSV file and set up a DateTime index
mens_rowing = pd.read_csv("datasets/mens_rowing.csv")
mens_gymnastics = pd.read_csv("datasets/mens_gymnastics.csv")
seattle_weather = pd.read_csv("datasets/seattle_weather_simple.csv")
austin_weather = pd.read_csv("datasets/austin_weather_simple.csv")
```

## Exercise: Adding error-bars to a bar chart

Statistical plotting can provide more insight by including measures of variability along with averages. In this exercise, you will visualize the mean height of medalists in two sports and add error bars to represent the standard deviation, giving a clearer comparison between the groups.

You are provided with two DataFrames:

* `mens_rowing` – contains information about medalists in rowing events.
* `mens_gymnastics` – contains information about medalists in gymnastics events.

### Instructions

1. Plot a bar for the mean height of `mens_rowing`, including an error bar showing its standard deviation.
2. Plot another bar for the mean height of `mens_gymnastics`, also with an error bar for the standard deviation.
3. Add a label to the y-axis: `"Height (cm)"`.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Add a bar for the rowing "Height" column mean/std
ax.bar("Rowing", mens_rowing["Height"].mean(), 
       yerr=mens_rowing["Height"].std())

# Add a bar for the gymnastics "Height" column mean/std
ax.bar("Gymnastics", mens_gymnastics["Height"].mean(), 
       yerr=mens_gymnastics["Height"].std())

# Label the y-axis
ax.set_ylabel("Height (cm)")

plt.show()
```

## Exercise: Adding error-bars to a plot

Error bars can also be applied to line plots, allowing us to display both the central trend and the variability in the data. In this task, you will use monthly weather data for two cities and add error bars to visualize the standard deviation of temperatures.

You are given two DataFrames:

* **`seattle_weather`** – contains columns `"MONTH"`, `"MLY-TAVG-NORMAL"`, and `"MLY-TAVG-STDDEV"`.
* **`austin_weather`** – structured in the same way for Austin.

### Instructions

1. Use `ax.errorbar` to plot Seattle’s monthly average temperatures with `"MONTH"` as x-values, `"MLY-TAVG-NORMAL"` as y-values, and `"MLY-TAVG-STDDEV"` as the error bars.
2. Use `ax.errorbar` again to plot Austin’s data in the same way.
3. Add the y-axis label: `"Temperature (Fahrenheit)"`.

```python
fig, ax = plt.subplots()

# Add Seattle temperature data with error bars
ax.errorbar(
    seattle_weather["MONTH"],
    seattle_weather["MLY-TAVG-NORMAL"],
    yerr=seattle_weather["MLY-TAVG-STDDEV"]
)

# Add Austin temperature data with error bars
ax.errorbar(
    austin_weather["MONTH"],
    austin_weather["MLY-TAVG-NORMAL"],
    yerr=austin_weather["MLY-TAVG-STDDEV"]
)

# Set the y-axis label
ax.set_ylabel("Temperature (Fahrenheit)")

plt.show()
```

## Exercise: Creating boxplots

Boxplots are a powerful way to summarize and compare distributions. They highlight the **median**, **interquartile range**, and the overall spread of the data. They also flag outliers that fall outside the typical range.

In this task, you will visualize and compare the heights of Olympic medalists from two sports using boxplots. The data is stored in two DataFrames:

* **`mens_rowing`** – contains a `"Height"` column for rowing medalists.
* **`mens_gymnastics`** – contains a `"Height"` column for gymnastics medalists.

### Instructions

1. Create a boxplot that displays the `"Height"` column from both DataFrames side by side.
2. Add tick labels on the x-axis: `"Rowing"` and `"Gymnastics"`.
3. Label the y-axis as `"Height (cm)"`.

```python
fig, ax = plt.subplots()

# Create boxplots for the heights of rowers and gymnasts
ax.boxplot([mens_rowing["Height"].dropna(), mens_gymnastics["Height"].dropna()])

# Add x-axis tick labels
ax.set_xticklabels(["Rowing", "Gymnastics"])

# Add a y-axis label
ax.set_ylabel("Height (cm)")

plt.show()
```
