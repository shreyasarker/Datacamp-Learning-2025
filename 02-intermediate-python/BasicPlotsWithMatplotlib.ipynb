## What is a Line Plot?

A **line plot** (also called a line chart or line graph) is a graphical representation of data points connected by straight line segments. It is commonly used to display trends or changes over time. The x-axis usually represents time or categories, while the y-axis represents the values that change over time or according to the categories.

In a line plot:
- Each data point corresponds to a pair of values (x, y).
- The data points are connected by a line, allowing us to see the overall trend.

Line plots are particularly useful for visualizing continuous data, such as population growth, stock prices, or temperature changes.


## Exercise: Line Plot (1)

In this exercise, you'll create a simple line plot using **matplotlib** to visualize the growth of the world population over time. A line plot helps you see trends and patterns in data across a range of values. 

Weâ€™ll be using data representing world population over the years, with `year` on the x-axis and `pop` on the y-axis. 

You will also practice printing the last values from both lists to see the estimated population for the year 2100.


```python
# Importing the plotting library
import matplotlib.pyplot as plt

# Example data for years and populations
year = [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2030, 2040, 2050, 2060, 2070, 2080, 2090, 2100]
pop = [2.5, 3.0, 3.7, 4.4, 5.3, 6.1, 6.9, 7.8, 8.5, 9.2, 9.8, 10.3, 10.7, 11.0, 11.3, 11.6]  # in billions

# Displaying the last items from both the year and population lists
print("Last year in the list:", year[-1])
print("Predicted population for the last year (2100):", pop[-1])
```

```python
# Creating the line plot: years on the x-axis, population on the y-axis
plt.plot(year, pop)

# Adding labels and a title to make the plot more readable
plt.xlabel('Year')
plt.ylabel('World Population (in billions)')
plt.title('World Population Over Time')

# Display the plot
plt.show()
```

## Exercise: Line Plot (2): Interpretation

Take another look at the plot you generated in the previous exercise, which is displayed on the right. According to the plot, around which year is the worldâ€™s population expected to exceed ten billion?

**Possible answers:**

```python
import numpy as np

# Example data for population growth
years = np.arange(2020, 2100, 1)
population = 7.8 * (1 + 0.01)**(years - 2020)  # An example exponential growth model

# Plotting the population growth
plt.plot(years, population, label='Population Growth')

# Adding a vertical line at the year 2060
plt.axvline(x=2060, color='r', linestyle='--', label="10 Billion Threshold")

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Population (Billions)')
plt.title('World Population Growth')

# Displaying the legend
plt.legend()

# Show the plot
plt.show()
```

## What is a Scatter Plot?

A **scatter plot** is a type of data visualization that displays values for two numerical variables for a set of data.

Each point in the scatter plot represents one observation from the dataset. The **horizontal axis (x-axis)** shows the value of one variable, and the **vertical axis (y-axis)** shows the value of the other.

Scatter plots are especially useful for:
- Visualizing the **relationship or correlation** between two variables.
- Identifying **patterns**, **trends**, or **outliers** in the data.
- Determining whether the relationship is **positive**, **negative**, or **nonexistent**.

For example, if you're analyzing **GDP per capita** vs **Life Expectancy**, each point would represent a country, with its GDP on the x-axis and its life expectancy on the y-axis.


## Exercise: Scatter Plot (1)

In situations where you need to assess the correlation between two variables, a scatter plot is often a more suitable choice than a line plot. A scatter plot displays individual data points, helping you understand if there is any trend or pattern between the variables.

For example, if we look at GDP per capita (GDP per person) and life expectancy across various countries, a scatter plot could help us visually identify if there's a relationship between these two variables.

In this exercise, youâ€™ll visualize the correlation between GDP per capita and life expectancy for different countries in 2007 using a scatter plot.

**Tasks:**
1. Modify the code to replace the line plot with a scatter plot.
2. To improve clarity and highlight the relationship, transform the x-axis (GDP per capita) to a logarithmic scale using `plt.xscale('log')`.
3. Finish by using `plt.show()` to display the plot.

Good luck visualizing the correlation between these variables!


```python
# Example data for GDP per capita (in USD) and life expectancy (in years)
gdp_per_person = [800, 1500, 2700, 3200, 4300, 5500, 6700, 8000, 9500, 11000]
life_expectancy_years = [42, 51, 56, 60, 63, 67, 71, 75, 78, 83]

# Create a scatter plot with GDP per capita on the x-axis and life expectancy on the y-axis
plt.scatter(gdp_per_person, life_expectancy_years)

# Use a logarithmic scale for the x-axis (GDP per capita) to better visualize the correlation
plt.xscale('log')

# Add labels and title to the plot
plt.xlabel('GDP per Capita (USD)')
plt.ylabel('Life Expectancy (Years)')
plt.title('Life Expectancy vs. GDP per Capita')

# Display the plot
plt.show()
```
