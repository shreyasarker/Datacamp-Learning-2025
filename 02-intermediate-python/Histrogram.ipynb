## What is a Histogram?

A **histogram** is a type of data visualization that shows the **distribution** of a single numerical variable.

It groups the data into **bins** (intervals), and for each bin, a **bar** represents how many data points fall into that range. This makes histograms ideal for understanding the **frequency** of values and identifying **patterns** such as:

- Skewness (left or right)
- Spread or range
- Clusters or gaps
- Outliers

**Key Points:**
- **X-axis:** Represents ranges (or bins) of values.
- **Y-axis:** Shows the number (frequency) of data points in each bin.
- **Each bar:** The height reflects how many values fall within that interval.

**Example:**
If you're analyzing the ages of people in a survey, a histogram could show how many people fall into the 0–10, 11–20, 21–30 age ranges, and so on.

Use histograms when you want to **see the shape** of the data distribution or **analyze how data is spread** across intervals.

## Exercise: Build a histogram (2): bins

In data visualization, a histogram allows us to observe how data is distributed across different intervals or **bins**.

In this exercise, we are working with the `life_exp` dataset, which contains the life expectancy for various countries in 2007.

### Experiment: Using Different Numbers of Bins

We will create two histograms:

1. **Histogram with 5 bins**: This groups the data into fewer, wider intervals.
2. **Histogram with 20 bins**: This gives us more granular insights by using narrower intervals.

### Observation

- **Fewer bins (5)** might hide important patterns by merging too many values.
- **More bins (20)** may reveal subtle variations but can look cluttered.

By comparing both, we can decide which representation better communicates the distribution of life expectancy across countries.


```python
# Imported matplotlib.pyplot as plt

# Defiined Life expectancy data

# Histogram with 5 bins
plt.hist(life_exp, bins=5, edgecolor='black')
plt.title("Life Expectancy - 5 Bins")
plt.xlabel("Life Expectancy")
plt.ylabel("Number of Countries")
plt.show()
plt.clf()

# Histogram with 20 bins
plt.hist(life_exp, bins=20, edgecolor='black')
plt.title("Life Expectancy - 20 Bins")
plt.xlabel("Life Expectancy")
plt.ylabel("Number of Countries")
plt.show()
plt.clf()
```
